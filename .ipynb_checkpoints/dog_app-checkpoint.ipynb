## Step 1: Detect Humans

In this section, we use OpenCV's implementation of [Haar feature-based cascade classifiers](http://docs.opencv.org/trunk/d7/d8b/tutorial_py_face_detection.html) to detect human faces in images.  

OpenCV provides many pre-trained face detectors, stored as XML files on [github](https://github.com/opencv/opencv/tree/master/data/haarcascades).  We have downloaded one of these detectors and stored it in the `haarcascades` directory.  In the next code cell, we demonstrate how to use this detector to find human faces in a sample image.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# Extract pre-trained face detector
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_alt.xml')

# Load color (BGR) image
img = cv2.imread(human_files[0])
# Convert BGR image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Find faces in image
faces = face_cascade.detectMultiScale(gray)

# Print number of faces detected in the image
print("Number of faces detected: ", len(faces))

# Get bounding box for each detected face
for (x, y, w, h) in faces:
    # add bounding box to color image
    cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 0), 2)
    
# Convert BGR image to RGB for plotting

cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image along with bounding box
plt.imshow(cv_rgb)
```

Before using any of the face detectors, it is standard procedure to convert the images to grayscale.  The `detectMultiScale` function executes the classifier stored in `face_cascade` and takes the grayscale image as a parameter.  

In the above code:
* `faces` is a numpy array of detected faces, where each row corresponds to a detected face.  
* Each detected face is a 1D array with four entries, `(x, y, w, h)`, that specifies the bounding box of the detected face.  
* The first two entries in the array (`x` and `y`) specify the **horizontal** and **vertical** positions of the <em>top left corner</em> of the bounding box.  
* The last two entries in the array (`w` and `h`) specify the **width** and **height** of the box.

### Human Face Detector

We can use this procedure to write a function that returns `True` if a human face is detected in an image and `False` otherwise. Function below, `face_detector`, takes a string-valued file path to an image as input and appears in the code block below.

```python
# Returns "True" if face is detected in image stored at img_path
def face_detector(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray)
    return len(faces) > 0
```

## Step 2: Detect Dogs

In this section, we use a [pre-trained model](http://pytorch.org/docs/master/torchvision/models.html) to detect dogs in images.  

### Obtain Pre-trained VGG-16 Model

The code below downloads the VGG-16 model, along with weights that have been trained on [ImageNet](http://www.image-net.org/), a very large, very popular dataset used for image classification and other vision tasks.  ImageNet contains over 10 million URLs, each linking to an image containing an object from one of [1000 categories](https://gist.github.com/yrevar/942d3a0ac09ec9e5eb3a).  

```python
import torch
import torchvision.models as models

# define VGG16 model
VGG16 = models.vgg16(pretrained=True)

# check if CUDA is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# move model to GPU if CUDA is available
VGG16 = VGG16.to(device)
```

Given an image, this pre-trained VGG-16 model returns a prediction (derived from the 1000 possible categories in ImageNet) for the object that is contained in the image.

###  Making Predictions with a Pre-trained Model

Below is a function that accepts a path to an image (such as `'dogImages/train/001.Affenpinscher/Affenpinscher_00001.jpg'`) as input and returns the index corresponding to the ImageNet class that is predicted by the pre-trained VGG-16 model.  The output should always be an integer between 0 and 999, inclusive.

Pre-processing tensors for pre-trained models: [PyTorch documentation](http://pytorch.org/docs/stable/torchvision/models.html).

```python
from PIL import Image
import torchvision.transforms as transforms

# Set PIL to be tolerant of image files that are truncated
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

def VGG16_predict(img_path):
    """ Use pre-trained VGG-16 model to obtain index corresponding to 
        predicted ImageNet class for image at specified path
        
        Params
        ------
        img_path(str): path to an image
        
        Returns
        -------
        Index corresponding to VGG-16 model's prediction
    """
    
    # Load and pre-process an image from the given img_path
    image = Image.open(img_path).convert('RGB')
    
    # Transform to cropped and normalized tensors
    transform = transforms.Compose([
                    transforms.Resize(224),
                    transforms.ToTensor(),
                    transforms.Normalize((0.485, 0.456, 0.406),
                                         (0.229, 0.224, 0.225))])
    image = transform(image).unsqueeze(0)
    image = image.to(device)
    # Return an index
    #vgg_out = VGG16(image)
    #_, pred = torch.max(vgg_out, 1)
    prob = torch.exp(VGG16(image))
    top_p, top_k = prob.topk(1)
    return top_k#pred.numpy()[0]
    
    
```

### Write a Dog Detector

Below is a function that returns `True` if detected image is a dog breed categorized in this [dictionary](https://gist.github.com/yrevar/942d3a0ac09ec9e5eb3a), of keys from 151 to 268 ( from `'Chihuahua'` to `'Mexican hairless'`).  Thus, in order to check to see if an image is predicted to contain a dog by the pre-trained VGG-16 model, we need only check if the pre-trained model predicts an index between 151 and 268 (inclusive).


```python
def dog_detector(img_path):
    ''' Returns "True" if a dog is detected in the image stored at img_path '''
    dog_idx = range(151, 269)
    if VGG16_predict(img_path) in list(dog_idx):
        return True
    else:
        return False
```

### Specify Data Loaders for the Dog Dataset

Create three separate [data loaders](http://pytorch.org/docs/stable/data.html#torch.utils.data.DataLoader) for the training, validation, and test datasets of dog images (located at `dogImages/train`, `dogImages/valid`, and `dogImages/test`, respectively).

* [Documentation on custom datasets](http://pytorch.org/docs/stable/torchvision/datasets.html) 
* [Documentation on augmenting/transforms](http://pytorch.org/docs/stable/torchvision/transforms.html?highlight=transform)

[Standard transforms usage for ImageNet reference](https://towardsdatascience.com/transfer-learning-with-convolutional-neural-networks-in-pytorch-dd09190245ce)

```python
import os
import requests
from torchvision import datasets

# Get data loaders for training, validation, and test sets

# Define training and test data directories
paths = ['dogImages/', 'train/', 'valid/', 'test/']
data_dirs = {paths[ii][:-1]: os.path.join(paths[0], paths[ii]) for ii in range(1, len(paths))}


# Function to augment and convert data to Tensors and get dataloaders
def get_dataloader(root, batch_size=128, shuffle=True, num_workers=4, set_type='train', 
                       img_size=256, crop_size=224, angle=15):

    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]
    
    transform_dict = {
        'train': transforms.Compose(
            [transforms.RandomResizedCrop(size=crop_size, scale=(0.8, 1.0)),
             transforms.RandomRotation(degrees=angle),
             transforms.ColorJitter(),
             transforms.RandomHorizontalFlip(),
             transforms.RandomVerticalFlip(),
             transforms.ToTensor(),
             transforms.Normalize(mean, std)
             ]),
        'valid': transforms.Compose(
            [transforms.Resize(size=img_size),
             transforms.CenterCrop(size=crop_size),
             transforms.ToTensor(),
             transforms.Normalize(mean, std)
             ]),
        'test': transforms.Compose(
            [transforms.Resize(size=img_size),
             transforms.CenterCrop(size=crop_size),
             transforms.ToTensor(),
             transforms.Normalize(mean, std)
             ])}

    dataset = datasets.ImageFolder(root, transform=transform_dict[set_type])
    dataloader = torch.utils.data.DataLoader(dataset, shuffle=shuffle, 
                                            batch_size=batch_size, num_workers=num_workers)
    return dataloader
        
# Create train, val and test data loader mapper
dataloaders = {x[:-1]: get_dataloader(data_dirs[x[:-1]], set_type=x[:-1])
                for x in paths[1:]}

# Dog class id list and dictionary to get corresponding breed names
dog_classes = {val: str(key)[4:] for key, val in dataloaders['train'].dataset.class_to_idx.items()}
```

### CNN Model Architecture

Create a CNN from scratch to classify dog breed.

[CNN architecture reference 1](https://colab.research.google.com/github/pratyakshajha/project-dog-classification/blob/master/dog_app.ipynb#scrollTo=t6Kpm-h9v2tU)

[ConvNet configuration](https://arxiv.org/pdf/1512.03385.pdf)


```python
import torch.nn as nn
import torch.nn.functional as F

# CNN architecture
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # Define a 2D CNN
        
    # conv layer 1 (sees 224x224x3 image tensor)
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
    # conv layer 2 (sees 112x112x112 tensor)
        self.conv2 = nn.Conv2d(16, 16, 3, padding=1)
    # conv layer 3 (sees 56x56x56 tensor)
        self.conv3 = nn.Conv2d(16, 32, 3, padding=1)
    # conv layer 4 (sees 28x28x28 tensor)
        self.conv4 = nn.Conv2d(32, 64, 3, padding=1)
    # Max pooling layer (2x2)
        self.pool = nn.MaxPool2d(2, 2)
        
    # fully connected layer(512 * 14 * 14 -> 4096)
        self.fc1 = nn.Linear(64*14*14, 4096)
    # fully connected layer(4096 -> 1000)
        self.fc2 = nn.Linear(4096, 1000)
    # final fully connected layer (133 classes out)
        self.fc3 = nn.Linear(1000, len(dog_classes))
    
    # dropout layer 
        self.dropout = nn.Dropout(0.5)
        
    def forward(self, x):
        ## Define forward behavior
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = self.pool(F.relu(self.conv4(x)))
        # Flatten the output of the last CNN layer
        x = x.view(-1, 64*14*14)
        # First dropout
        x = self.dropout(x)
        # First 2 fc with relu, followed by dropouts
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        # Final fc layer itself
        x = self.fc3(x)
        
        return x

#-#-# You do NOT have to modify the code below this line. #-#-#

# instantiate the CNN
model_scratch = Net()

# move tensors to GPU if CUDA is available
model_scratch = model_scratch.to(device)
```

__Question 4:__ Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  

__Answer:__ Based on what I learned from looking at various CNN architectures and the configuration descriptions through research papers, (as well as my own image inputs), using 3-5 deep conv layers would make sense. Starting with the first layer with 3 dimensional (rgb) intakes and 16 outputs with padding of 1 (stride at default) to match the dimension of the image input. I thought it would be a good idea to use batch norm, but I tried using simply a layer with same number outputs in the second layer. I continued building up to 4th conv layer with 64 outputs, so that I can match the computation of input dimension in the first fully connected layer to be greater than 4096, which can be seen in the in_features of VGG16 classifier network. I completed the model with 3 fully connected layers, in which the last layer does not get dropouts and relu activation. All Conv layers had max pooling of 2x2 as stated in the paper references. 

###  Specify Loss Function and Optimizer

Specify a [loss function](http://pytorch.org/docs/stable/nn.html#loss-functions) and [optimizer](http://pytorch.org/docs/stable/optim.html).  Save the chosen loss function as `criterion_scratch`, and the optimizer as `optimizer_scratch`.

```python
import torch.optim as optim
from torch.optim import lr_scheduler

# CrossEntropyLoss as loss function
criterion_scratch = nn.CrossEntropyLoss()
#criterion_scratch = criterion_scratch.to(device)

# Adam optimizer
optimizer_scratch = optim.Adam(model_scratch.parameters(), lr=0.0003, betas=[0.5, 0.999])

# Learning rate decay scheduler
scheduler_scratch = lr_scheduler.StepLR(optimizer_scratch, step_size = 10, gamma=0.1)
```

### Train and Validate the Model

[Save the final model parameters](http://pytorch.org/docs/master/notes/serialization.html) at filepath `'model_scratch.pt'`.

```python
# Good idea to record the stats for later reference
loss_dict = {'epoch': [],
          'train_losses': [],
          'val_losses': [],
          'test_losses': []}

acc_dict = {'epoch': [],
            'train_acc': [],
            'val_acc': [],
            'test_acc': []}
```

```python
# The following import is required for training to be robust to truncated images
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True

def train(epochs, loaders, model, optimizer, criterion, device, save_path):
    """returns trained model"""
    # initialize tracker for minimum validation loss
    val_loss_min = np.Inf 
    
    for ep in range(1, epochs+1):
        loss_dict['epoch'].append(ep)
        acc_dict['epoch'].append(ep)
        # initialize variables to monitor training and validation loss
        train_loss = 0.0
        val_loss = 0.0
        train_correct = 0
        val_correct = 0
        
        ###################
        # train the model #
        ###################
        model.train()
        for batch_idx, (data, target) in enumerate(loaders['train']):
            # move to GPU
            data, target = data.to(device), target.to(device)
            # Clear batch gradients
            optimizer.zero_grad()
            # Forward pass
            outputs = model(data)
            pred = outputs.data.max(1, keepdim=True)[1]
            # Get batch loss
            loss = criterion(outputs, target)
            # Back prop
            loss.backward()
            optimizer.step()
            
            # Update average training loss and total correct counts
            train_loss += loss.item() * data.size(0)
            train_correct += torch.sum(pred.eq(target.data.view_as(pred)))   
            
        ######################    
        # validate the model #
        ######################
        model.eval()
        for batch_idx, (data, target) in enumerate(loaders['valid']):
            # Move to GPU
            data, target = data.to(device), target.to(device)
            # Forward pass
            outputs = model(data)
            pred = outputs.data.max(1, keepdim=True)[1]
            # Calculate batch loss
            loss = criterion(outputs, target)
            
            # Update the average val loss and total correct counts
            val_loss += loss.item() * data.size(0)
            val_correct += torch.sum(pred.eq(target.data.view_as(pred)))   #torch.sum(pred == target.data)
        
        # Get mean loss and add to losses dict
        train_loss  = train_loss/len(loaders['train'].dataset)
        val_loss = val_loss/len(loaders['valid'].dataset)
        loss_dict['train_losses'].append(train_loss)
        loss_dict['val_losses'].append(val_loss)
        
        # Add the average acc into acc dict
        train_acc = train_correct.double()/len(loaders['train'].dataset)
        val_acc = val_correct.double()/len(loaders['valid'].dataset)
                                               
        acc_dict['train_acc'].append(train_acc)
        acc_dict['val_acc'].append(val_acc)

        # print training/validation statistics 
        print('Epoch: {} |Train Loss: {:.6f} | Val Loss: {:.6f} | train acc/val acc: {:.2%}/{:.2%}'.format(
            ep, 
            train_loss,
            val_loss,
            train_acc,
            val_acc
            ))
        
        # Save the model if validation loss has decreased
        if val_loss <= val_loss_min:
            print("# --- Validation loss decreased. Saving model... --- #\n")
            torch.save(model.state_dict(), save_path)
            val_loss_min = val_loss
            
    # return trained model
    return model


# train the model
model_scratch = train(30, dataloaders, model_scratch, optimizer_scratch, 
                      criterion_scratch, device, 'model_scratch.pt')

# load the model that got the best validation accuracy
model_scratch.load_state_dict(torch.load('model_scratch.pt'))
```

### Test the Model

Try out the model on the test dataset of dog images.  Use the code cell below to calculate and print the test loss and accuracy.  Shoot for > 10%.

```python
def test(loaders, model, criterion, use_cuda):

    # monitor test loss and accuracy
    test_loss = 0.
    correct = 0.
    total = 0.

    model.eval()
    for batch_idx, (data, target) in enumerate(loaders['test']):
        # move to GPU
        data, target = data.to(device), target.to(device)
        # forward pass: compute predicted outputs by passing inputs to the model
        output = model(data)
        # calculate the loss
        loss = criterion(output, target)
        # update average test loss 
        test_loss += loss.item()*data.size(0)
        # convert output probabilities to predicted class
        pred = output.data.max(1, keepdim=True)[1]
        # compare predictions to true label
        correct += np.sum(np.squeeze(pred.eq(target.data.view_as(pred))).cpu().numpy())
        total += data.size(0)
            
    print('Test Loss: {:.6f}\n'.format(test_loss))

    print('\nTest Accuracy: %2d%% (%2d/%2d)' % (
        100. * correct / total, correct, total))

# call test function    
test(dataloaders, model_scratch, criterion_scratch, device)

```

### CNN Model Architecture with Transfer Learning

Use transfer learning to create a CNN to classify dog breed.  Initialize model as the variable `model_transfer`.

[ResNet152 Model Architecture Reference](https://arxiv.org/pdf/1512.03385.pdf)

```python
import torchvision.models as models
import torch.nn as nn
from torch.autograd import Variable
from collections import OrderedDict

# Using ResNet152 pretrained model
model_transfer = models.resnet152(pretrained=True)
# Pop off the last layer of the model and save the name and features
classifier, old_classifier = model_transfer._modules.popitem()

# Freeze the parameters
for param in model_transfer.parameters():
    param.requires_grad = False

# Save the input and output sizes in the old classifier layers
input_size = old_classifier.in_features
output_size = old_classifier.out_features

# Define a new classifier layer to add
new_classifier = nn.Sequential(OrderedDict([
                           ('fc1', nn.Linear(input_size, len(dog_classes))),
                           ('output', nn.LogSoftmax(dim=1))
                           ]))
"""new_classifier = nn.Sequential(OrderedDict([
                           ('fc1', nn.Linear(input_size, output_size)),
                           ('relu', nn.ReLU()),
                           ('dropout', nn.Dropout(p=0.5)),
                           ('fc2', nn.Linear(output_size, len(dog_classes))),
                           ('output', nn.LogSoftmax(dim=1))
                           ]))"""
# Add the new sequences of layers to the model
model_transfer.add_module(classifier, new_classifier)
# To GPU if available
model_transfer = model_transfer.to(device)
# View the model layers
print(model_transfer)
```

__Question 5:__ Outline the steps you took to get to your final CNN architecture and your reasoning at each step.  Describe why you think the architecture is suitable for the current problem.

__Answer:__ I read a paper on ResNet152, which seemed to yield pretty low losses compared to VGG16. It has a very deep architecture, with batch norm uses, which I prefer. I only included one linear layer at the final classifer, after removing the original classifier layer of resnet152 model. I used the input feature size as the original input size, and output feature size as the number of dog breeds in our dataset. Finally, I applied LogSoftmax to the output to convert the out value from percentage.

### Predict Dog Breed with the Model

Below is a function that takes an image path as input and returns the dog breed (`Affenpinscher`, `Afghan hound`, etc) that is predicted by the new CNN model.  

```python
def predict_breed_transfer(img_path):
    """ Takes in an image path to read and convert as appropriate input to
        the CNN model. Returns the name of a dog breed that model predicts. """
    # load the image and return the predicted breed
    image = Image.open(img_path).convert('RGB')
    
    # Transform to cropped and normalized tensors
    transform = transforms.Compose([
                    transforms.Resize(224),
                    transforms.ToTensor(),
                    transforms.Normalize((0.485, 0.456, 0.406),
                                         (0.229, 0.224, 0.225))])
    # Get top label
    image = transform(image).unsqueeze(0)
    image = image.to(device)
    prob = torch.exp(model_transfer(image))
    top_p, top_k = prob.topk(1)
    
    return dog_classes[int(top_k)]
```
